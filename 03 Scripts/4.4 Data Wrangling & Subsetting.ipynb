# Creating a new header and using the "iloc[ ]" function 

Used for selecting data in pandas and takes the index number of a row or column (or multiple rows and columns).By executing and assigning it to a variable, the values within that row(s) or column(s)are assigned to that variable.

```python
# tell Python to take the row under the current header (the first row in the df)
# use that data to create a new header, and get rid of the current header
new_header = df_dep_t.iloc[0]
```

```python
# check
new_header
```

```python
# create a new dataframe that only copies over rows beyond the first row
# Take the data under the header row for a new df
df_dep_t_new = df_dep_t[1:]
```

What does that 1: mean? This is telling Python to copy everything from the first row onward from the df_dep_t dataframe into a new dataframe, df_dep_t_new. The 1 represents row 1 in the dataframe. Remember that since Python starts indexing from 0 rather than 1, “row 1” is actually the second row in the dataframe. The colon that follows the 1 represents onward until the end of the dataframe. Together, they tell Python to copy everything starting from row 1 until the end of the dataframe.

```python
df_dep_t_new
```

```python
# set your new_header variable as the new header for your df_dep_t_new dataframe
# tells Python to use the list of column names in your new_header variable as your new header
df_dep_t_new.columns = new_header
```

```python
df_dep_t_new
```
